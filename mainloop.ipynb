```python
import time
from datetime import datetime
import json
from constants import *

import behavior   # behavior.ipynb
import visualizer as vz # Tkinter.ipynb

reload(behavior)
reload(vz)
reload(constants)

# Update rates (arbitrary units)



def run_simulation(decider, actor_data, circle_map, track, to_file=None):
    decider.set_map(circle_map)
    dt = vz.SECS_PER_LOOP
    frame_num = 0
    output = {
        "ego" : {
            "id" : "ego",
            "length" : 6.0,
            "width" : 6.0,
            "predictions" : []
                
        }
    }
    while frame_num < TOTAL_FRAMES:
        data = {
            "s" : decider.ego.s,
            "d" : decider.ego.d,
            "v" : decider.ego.speed,
            "timestamp": frame_num*dt,
            "lane": decider.ego.lane
        }
        output["ego"]["predictions"].append(data)
        if frame_num % BEHAVIOR_UPDATE_RATE == 0:
            old_state = decider.state
            new_state = decider.get_behavior()
            if new_state != old_state:
                print
                print "Changing from {} to {} in frame {}".format(old_state, new_state, frame_num)
                print
            v = decider.get_velocity_for_state(new_state)
            if new_state.startswith("lane_change"):
                diffs = {"lane_change_left":-1, "lane_change_right":+1}
                other_lane = decider.ego.lane + diffs[new_state]
                ds, dd = decider.map.distance_between(decider.ego.pose, behavior.Pose(decider.ego.s, 0, other_lane))
                print "ds is {}".format(ds)
#                 decider.ego.set_lane(other_lane)
                decider.ego.set_pose(behavior.Pose(decider.ego.s - ds, 0, other_lane))
            decider.ego.set_velocity(v)
        decider.ego.update(dt, track)
        frame_num += 1
        for v_id, v in actor_data["vehicles"].items():
            v["predictions"].pop(0)
    if not to_file:
        return
    with open(to_file, 'wb') as f:
        json.dump(output, f) 
    

def main():
    start = datetime.now()
    with open(ACTOR_DATA_FILE_NAME, 'rb') as f:
        actor_data = json.load(f)
        
    ego = behavior.Vehicle(5.0, 5.0)
    P = INITIAL_EGO_POSE
    ego.set_pose(behavior.Pose(P[0], P[1], P[2]))
    ego.set_velocity(30.0)
    decider = behavior.Decider(ego)
    G = GOAL
    decider.set_goal(behavior.Location(G[0], G[1]))
    decider._update_with_new_predictions(actor_data)
    
    output_file_name = "simulation.json"
    run_simulation(decider, actor_data, behavior.CircleMap(TRACK_RADIUS,NUM_LANES,LANE_WIDTH), 
                   vz.Track(TRACK_RADIUS,NUM_LANES,LANE_WIDTH), to_file=output_file_name)
    
    
        
    
    done_with_simulation = datetime.now()
    simulation_duration = done_with_simulation - start
    print "took {} to run simulation".format(simulation_duration.total_seconds())
    with open(output_file_name, 'rb') as f:
        
        data = json.load(f)
    visualize_result(data)
    
    


def visualize_result(data):
    print "VISUALIZE"
    s = vz.Simulation()
    track = vz.Track(TRACK_RADIUS,NUM_LANES,LANE_WIDTH)
    s.draw_track(track, vz.CANVAS_WIDTH, vz.CANVAS_HEIGHT)
    s.add_cars(s.car_size, 0, CARS_PER_LANE[0])
    s.add_cars(s.car_size, 1, CARS_PER_LANE[1])
    s.animate(with_ego_data=data)
#     with open("output.json", 'wb') as f:
#         json.dump(s.output, f)
    pass
```
